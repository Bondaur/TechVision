Задание №1

```python
import cv2
import numpy as np

def morph_conv(img, op, kernel, iterations = 1):
    original = img
    match op:
        case cv2.MORPH_CLOSE:
            dilated_image = cv2.dilate(original,kernel,iterations=iterations)
            eroded_image = cv2.erode(dilated_image,kernel,iterations=iterations)
            return eroded_image
        case cv2.MORPH_OPEN:
            eroded_image = cv2.erode(original,kernel,iterations=iterations)
            dilated_image = cv2.dilate(eroded_image,kernel,iterations=iterations)
            return dilated_image
        
path = r"2-1.jpg"
picture = cv2.imread(path,cv2.IMREAD_COLOR)

kernel = np.ones((5,5), np.uint8)

user_opened_picture = morph_conv(picture,cv2.MORPH_OPEN,kernel,iterations = 1)
sys_opened_picture = cv2.morphologyEx(picture, cv2.MORPH_OPEN, kernel, iterations = 1)
open_diff = sys_opened_picture - user_opened_picture

user_closed_picture = morph_conv(picture,cv2.MORPH_CLOSE,kernel, iterations = 1)
sys_closed_picture = cv2.morphologyEx(picture, cv2.MORPH_CLOSE, kernel, iterations = 1)
close_diff = sys_closed_picture - user_closed_picture

cv2.imshow("2-1.jpg",picture)
cv2.imshow("sys_opened_picture.jpg",sys_opened_picture)
cv2.imshow("user_open.jpg",user_opened_picture)
res = cv2.imwrite(r"result_1.jpg",user_opened_picture)
cv2.imshow("user_close.jpg",user_closed_picture)
cv2.imshow("sys_closed_picture.jpg",sys_closed_picture)
cv2.imshow("open_diff.jpg",open_diff)
cv2.imshow("close_diff.jpg",close_diff)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
path = r"2-2.jpg"
picture = cv2.imread(path,cv2.IMREAD_GRAYSCALE)

kernel = np.ones((3,3), np.uint8)

sys_opened_picture = cv2.morphologyEx(picture,cv2.MORPH_OPEN,kernel=kernel,iterations=1)
sys_blured_opened_picture = cv2.medianBlur(sys_opened_picture,3)

cv2.imshow("2-2.jpg",picture)
cv2.imshow("sys_opened_picture",sys_opened_picture)
cv2.imshow("sys_blured_opened_picture",sys_blured_opened_picture)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
