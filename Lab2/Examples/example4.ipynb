### Морфологические преобразования

```python
import cv2
import numpy as np
```

Поговорим о двух мощных инструментах для морфологической обработки изображений - эрозии и наращивании. Эти операции помогают нам уточнять формы объектов на изображении, удаляя шум и соединяя разделенные элементы.

Эрозия уменьшает размер объектов, удаляя пиксели по краям, в то время как дилатация увеличивает размер объектов, добавляя пиксели к границам.

```python

image = cv2.imread('image.png',cv2.IMREAD_REDUCED_COLOR_4)
kernel = np.ones((3,3), np.uint8)

# Применение эрозии
eroded_image = cv2.erode(image, kernel, iterations=2)




```

```python
# Применение наращивания
dilated_image = cv2.dilate(image, kernel, iterations=2)
```

На основе наращивания и эрозии работают другие варианты морфологиче-
ских преобразований. Их может быть также полезно применять в случае, когда

исходное изображение рассматривается не как бинарное, а как полутоновое (в от-
тенках серого). Рассмотрим наиболее популярные операции. Все они так или

иначе будут сочетаниями базовых, то есть наращивания и эрозии. И все они могут
быть применены с помощью одной функции OpenCV – morphologyEx().

Все аргументы функции, за исключением двух, совпадают с аргументами
функций dilate() и erode(). Обязательно нужно задать входное изображение и
содержание ядра, а положение якорной точки, число итераций, тип рамки и ее
значение можно оставить по умолчанию. Если не указывать число итераций, то
преобразование будет применяться один раз.
Еще один обязательный аргумент – op – тип операции. Именно он
определяет, какое преобразование будет выполняться. Данная функция
позволяет применить ряд преобразований, среди которых мы рассмотрим
следующие:

cv2.MORPH_OPEN – размыкание;

cv2. MORPH _CLOSE – замыкание;



Теперь разберем, как выполняются разные типы операций. Начнем с размы-
кания. Оно выполняется в два шага. Сначала к исходному результату применя-
ется операция эрозии. Затем к результату эрозии применяется наращивание.

```python
img_open = cv2.morphologyEx(image, cv2.MORPH_OPEN, kernel, iterations=1)
```

Парная к размыканию операция – это замыкание. Отличие заключается в
том, что в замыкании сначала выполняется наращивание, и только затем – эрозия.


```python
img_close = cv2.morphologyEx(image, cv2.MORPH_CLOSE, kernel, iterations=1)
```

```python
cv2.imshow('Original', image)
cv2.imshow('Erosion', eroded_image)
cv2.imshow('Dilation', dilated_image)
cv2.imshow('Open', img_open)
cv2.imshow('Close', img_close)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
