Задание №1, в качестве изображения - фото 1-ого корпуса ЛЭТИ

```python
import cv2

#1
path = r"./1.png"
cv2.namedWindow("1")
picture = cv2.imread(path,cv2.IMREAD_COLOR)
cv2.imshow("1",picture)
cv2.waitKey(5000)
cv2.destroyWindow("1")
#2
picture2 = cv2.cvtColor(picture,cv2.COLOR_BGR2GRAY)
cv2.imshow("2",picture2)
cv2.waitKey(7000)
cv2.destroyWindow("2")
#3
cv2.namedWindow("3")
picture3 = cv2.imread(path,cv2.IMREAD_REDUCED_COLOR_2)
cv2.imshow("3",picture3)
cv2.waitKey(9000)
cv2.destroyWindow("3")
#4
cv2.namedWindow("4")
picture4 = cv2.imread(path,cv2.IMREAD_REDUCED_GRAYSCALE_4)
cv2.imshow("4",picture4)
cv2.waitKey(11000)
cv2.destroyWindow("4")
#5
cv2.namedWindow("5")
b, g, r = cv2.split(picture)
picture5 = cv2.merge([b, r, g])
cv2.imshow("5", picture5)
cv2.waitKey(4000)
cv2.destroyWindow("5")
#6
cv2.namedWindow("6")
picture6 = cv2.imread(path,cv2.IMREAD_REDUCED_COLOR_4)
cv2.imshow("6",picture6)
while(True):
    pressed_key = cv2.waitKey(0)
    if pressed_key == 27: break
cv2.destroyWindow("6")
```

Задание №2

```python

path = r"1-1.jpg"
image = cv2.imread(path, cv2.IMREAD_REDUCED_GRAYSCALE_2)

# Определение порога
threshold_value = 100

# Бинарное пороговое преобразование
_, binary = cv2.threshold(image, threshold_value, 255, cv2.THRESH_BINARY)

# Инвертированное бинарное пороговое преобразование
_, binary_inv = cv2.threshold(image, threshold_value, 255, cv2.THRESH_BINARY_INV)

# Пороговое преобразование с усечением
_, trunc = cv2.threshold(image, threshold_value, 255, cv2.THRESH_TRUNC)

# Пороговое преобразование до нуля
_, tozero = cv2.threshold(image, threshold_value, 255, cv2.THRESH_TOZERO)

# Инвертированное пороговое преобразование до нуля
_, tozero_inv = cv2.threshold(image, threshold_value, 255, cv2.THRESH_TOZERO_INV)

# метод Отсу
tr1, otsu = cv2.threshold(image, threshold_value, 255, cv2.THRESH_OTSU)
# Метод треугольника
tr2, triangle = cv2.threshold(image, threshold_value, 255, cv2.THRESH_TRIANGLE)

adaptive_thresh = cv2.adaptiveThreshold(image, 255, cv2.ADAPTIVE_THRESH_MEAN_C,
                                        cv2.THRESH_BINARY, 9, 15)

cv2.imshow('Original', image)
cv2.imshow('Binary', binary)
cv2.imshow('Binary Inverted', binary_inv)
cv2.imshow('Truncate', trunc)
cv2.imshow('To Zero', tozero)
cv2.imshow('To Zero Inverted', tozero_inv)


cv2.imshow('otsu', otsu)
cv2.imshow('triangle', triangle)
cv2.imshow('Adaptive Threshold', adaptive_thresh)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

Задание №5

```python
path = r"1-4.png"
picture = cv2.imread(path,cv2.IMREAD_REDUCED_COLOR_2)


blur = cv2.blur(picture, (11,11))
boxFiltered = cv2.boxFilter(picture, -1, (21, 21), normalize=True)
median = cv2.medianBlur(picture,21)
gaus = cv2.GaussianBlur(picture, (9, 9), 10)
bilat = cv2.bilateralFilter(picture,100,100,150)

cv2.imshow("1-4.png",picture)
cv2.imshow("blur",blur)
cv2.imshow("boxFiltered",boxFiltered)
cv2.imshow("median",median)
cv2.imshow('Gaussian Blurred', gaus)
cv2.imshow('bialt', bilat)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
