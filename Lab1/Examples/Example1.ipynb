### Изменение цветовой модели

```python
# Преобразование в серый
path='1.png'
picture=cv2.imread(path, cv2.IMREAD_REDUCED_COLOR_2)
gray_image = cv2.cvtColor(picture, cv2.COLOR_BGR2GRAY)
cv2.imshow("Test_img",picture)
cv2.imshow('Gray Image', gray_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Преобразование в HSV
hsv_image = cv2.cvtColor(picture, cv2.COLOR_BGR2HSV)
cv2.imshow("Test_img",picture)
cv2.imshow('HSV Image', hsv_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### Меняем каналы местами

```python
path='2.png'
picture=cv2.imread(path, cv2.IMREAD_REDUCED_COLOR_2)
b, g, r = cv2.split(picture)
cv2.imshow("Test_img",picture)
cv2.imshow("b", b)
cv2.imshow("g", g)
cv2.imshow("r", r)

img_rgb = cv2.merge([r, g, b])
cv2.imshow("merged", img_rgb)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### Рисование на изображениях

```python
empty_color_image = np.full((400,400,3),[50,50,100],np.uint8)
line_image = empty_color_image.copy()
cv2.line(line_image, (50, 50), (250, 250), (255, 0, 0), 5)  # Синяя линия
cv2.imshow('Line Image', line_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Отрисовка круга
circle_image = empty_color_image.copy()
cv2.circle(circle_image, (150, 150), 50, (0, 255, 0), 2)  # Зеленый круг
cv2.imshow('Circle Image', circle_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Отрисовка прямоугольника
rectangle_image = empty_color_image.copy()
cv2.rectangle(rectangle_image, (50, 50), (250, 250), (0, 0, 255), 5)  # Красный прямоугольник
cv2.imshow('Rectangle Image', rectangle_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Отрисовка текста
text_image = empty_color_image.copy()
cv2.putText(text_image, 'OpenCV', (70, 150), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 255, 255), 3)  # Белый текст
cv2.imshow('Text Image', text_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python

sheet = np.full((400, 500, 3),255,np.uint8)
cv2.line(sheet,(1,1),(500,400),(0,255,0),3)
cv2.putText(sheet,'line',(60,30),cv2.FONT_HERSHEY_SCRIPT_SIMPLEX,1,(0,0,0))
cv2.circle(sheet,(400,100),50,(255,0,0),3)
cv2.putText(sheet,'circle',(280,100),cv2.FONT_HERSHEY_SCRIPT_SIMPLEX,1,(0,0,0))
cv2.rectangle(sheet,(50,300),(150,350),(0,0,255),3)
cv2.putText(sheet,'rectangle',(50,280),cv2.FONT_HERSHEY_SCRIPT_SIMPLEX,1,(0,0,0))
cv2.imshow('print', sheet)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
