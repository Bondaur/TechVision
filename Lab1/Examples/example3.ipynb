```python
import cv2
import numpy as np
```

Функция cv2.medianBlur использует медианный фильтр для сглаживания изображения. В отличие от размытия по Гауссу, этот метод заменяет каждый пиксель на медиану значений в его окрестности. Это особенно полезно для удаления шумов, так как медиана устойчива к выбросам.

```python
path="2-1.jpg"
img=cv2.imread(path,cv2.IMREAD_REDUCED_COLOR_4)
imgnew=cv2.medianBlur(img,9)
cv2.imshow("res",imgnew)
cv2.imshow("original",img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

Двусторонний фильтр считает для каждого пикселя средневзвешенное значение в зависимости от яркости пикселя и его соседей. Важное отличие дву-
стороннего фильтра от фильтра Гаусса заключается в его весовой функции.

Первый компонент весовой функции двустороннего фильтра учитывает расстояние между центральным пикселем и текущим соседним. Второй компо-
нент основан на разнице между яркостью центрального пикселя и данного соседнего. Такой фильтр может быть полезен при подготовке изображения к сег-
ментации. В OpenCV двусторонний фильтр реализован в функции bilaterialFilter().

```python
path="3-3.jpg"
img=cv2.imread(path,cv2.IMREAD_REDUCED_COLOR_4)
imgnew=cv2.medianBlur(img,9)
cv2.imshow("res",imgnew)
cv2.imshow("original",img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
path="2-1.jpg"
img=cv2.imread(path,cv2.IMREAD_REDUCED_COLOR_4)
imgnew=cv2.medianBlur(img,9)
imgb=cv2.bilateralFilter(img,100,100,100)
cv2.imshow("res",imgnew)
cv2.imshow("original",img)
cv2.imshow("bilater",imgb)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
