### Границы

```python
import cv2
import numpy as np
```

```python
image = cv2.imread('image.png', cv2.IMREAD_REDUCED_GRAYSCALE_4)
# image = cv2.imread('theresh.png', cv2.IMREAD_GRAYSCALE)
_, image = cv2.threshold(image, 200, 255, cv2.THRESH_BINARY_INV)
```

Определение границ — это процесс выявления резких переходов интенсивности или цвета на изображении. Другими словами, представьте себе переход от одной текстуры к другой, от светлого к темному — это и есть границы, которые мы стремимся обнаружить. 

В OpenCV существует несколько подходов к вычислению границ – это cv2.Sobel(), cv2.Canny() и cv2.Laplacian(). 


```python
# Применение оператора Собеля для выявления горизонтальных и вертикальных границ
sobelx = cv2.Sobel(image, cv2.CV_64F, 1, 0, ksize=11)  # Градиент по X
sobely = cv2.Sobel(image, cv2.CV_64F, 0, 1, ksize=5)  # Градиент по Y
laplace = cv2.Laplacian(image, -1, ksize=3)
```

```python
# Применение детектора границ Кенни с заданными значениями градиента яркости изображения в направлении x и y
edges = cv2.Canny(image, 100, 200)
```

```python
cv2.imshow('Original', image)
cv2.imshow('x', sobelx)
cv2.imshow('y', sobely)
cv2.imshow('canny', edges)
cv2.imshow('laplician', laplace)

cv2.waitKey(0)
cv2.destroyAllWindows()
cv2.waitKey(0)
cv2.destroyAllWindows()
```
