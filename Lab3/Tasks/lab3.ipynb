Задание №1

```python
import cv2
import numpy as np
```

```python
path = r"./3-1.jpg"
picture = cv2.imread(path, cv2.IMREAD_REDUCED_GRAYSCALE_8)
sobelx = cv2.Sobel(picture, cv2.CV_64F, 1, 0, ksize=11)  # Градиент по X
sobely = cv2.Sobel(picture, cv2.CV_64F, 0, 1, ksize=5)  # Градиент по Y
laplace = cv2.Laplacian(picture, -1, ksize=3)
cv2.imshow("3-1.jpg",picture)
cv2.imshow("x",sobelx)
cv2.imshow("y",sobely)
cv2.imshow("laplace",laplace)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

Задание №2

```python
path = r"./3-4.png"
picture = cv2.imread(path, cv2.IMREAD_REDUCED_GRAYSCALE_2)
median = cv2.imread(r"./median.jpg",cv2.IMREAD_GRAYSCALE)
bilat = cv2.imread(r"./bilat.jpg",cv2.IMREAD_GRAYSCALE)
edges = cv2.Canny(picture, 100, 200)
median_edges = cv2.Canny(median, 36, 36)
bilat_edges = cv2.Canny(bilat, 100, 100)
cv2.imshow("median",median)
cv2.imshow("bilat",bilat)
cv2.imshow("3-4.png",picture)
cv2.imshow('canny', edges)
cv2.imshow('median_edges', median_edges)
cv2.imshow('bilat_edges', bilat_edges)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
